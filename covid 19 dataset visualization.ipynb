# 12. Covid Data visualization using India map and packages for data visualization

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("state_wise.csv")
final=df[['State','Confirmed','Recovered','Deaths','Active']]
final=final.drop([0])
final
```

```python
plt.hist(final['Deaths'])
plt.xlabel('Deaths')
plt.ylabel('State count')
```

```python
new_df=final[:5]
```

```python
new_df
```

```python
import numpy as np
```

```python
new_df.plot(kind='area')
plt.xlabel('States')
plt.ylabel('Cases')
plt.show()
```

```python
count,bin_edges=np.histogram(final['Deaths'])

final['Deaths'].plot(kind='hist',xticks=bin_edges)
```

```python
plt.figure(figsize=(10,7))

plt.pie(final['Confirmed'] ,labels=final['State'],autopct='%1.1f%%',radius=1.5)
plt.show()

# Maharashtra is having maximum confirmed cases
```

```python
plt.figure(figsize=(10,7))

plt.pie(final['Deaths'] ,labels=final['State'],autopct='%1.1f%%',radius=1.5)
plt.show()

# Maharashtra is having maximum confirmed cases
```
